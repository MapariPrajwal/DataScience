```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_edges_from([
    ('Alice', 'Bob'),
    ('Alice', 'Carol'),
    ('Alice', 'David'),
    ('Bob', 'Carol'),
    ('Carol', 'David'),
    ('David', 'Eve'),
    ('Eve', 'Frank')
])
```

```python
def create_ego_network(graph, ego_node, radius=1):
    ego_network = nx.ego_graph(graph, ego_node, radius=radius)
    return ego_network
ego_node = 'Alice'

ego_network = create_ego_network(G, ego_node, radius=1)

plt.figure(figsize=(8, 8))
pos = nx.spring_layout(ego_network)
nx.draw(ego_network, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=12)
plt.title("Ego Network of {}".format(ego_node))
plt.show()
```
